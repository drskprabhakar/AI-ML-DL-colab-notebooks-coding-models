```python
y = df['glaucoma'].values
x = df.drop(['glaucoma'], axis=1).values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# Assuming x and y are defined somewhere above
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

modellog = LogisticRegression()
# Fit the model using the training data
modellog.fit(x_train, y_train)

# Predict using the test data
y_predlog = modellog.predict(x_test)

# Now, calculate the confusion matrix
metrics.confusion_matrix(y_test, y_predlog)
```

```python
from sklearn.linear_model import LogisticRegression
modellog = LogisticRegression()
modellog.fit(x,y)
```

```python
# Reshape x_test to be a 2D array before prediction
y_predlog = modellog.predict(x_test.reshape(-1, 1))
y_predlog
```

```python
modellog.score(x,y)
```

```python
print(metrics.classification_report(y_test, y_predlog))
```
