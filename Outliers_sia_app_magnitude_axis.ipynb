```python
q1m=df.magnitude.quantile(0.25)
q3m = df.magnitude.quantile(0.75)
q1m,q3m
```

```python
iqrm = q3m-q1m
iqrm
```

```python
lowerlimitm = q1m-1.5+iqrm
upperlimitm = q3m+1.5*iqrm
lowerlimitm, upperlimitm
```

```python
df1 = df[(df.magnitude<upperlimitm) & (df.magnitude>lowerlimitm)]
df1.shape
```
