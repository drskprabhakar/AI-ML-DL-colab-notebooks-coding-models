```python
import matplotlib.pyplot as plt

# Given values
thresholds = [0.0, 0.25, 0.5, 0.75, 1.0]
TP = [25, 22, 18, 10, 0]
FP = [25, 9, 5, 2, 0]
TN = [0, 16, 20, 23, 25]
FN = [0, 3, 7, 15, 25]

# Calculate TPR and FPR
TPR = [tp / (tp + fn) for tp, fn in zip(TP, FN)]
FPR = [fp / (fp + tn) for fp, tn in zip(FP, TN)]

# Plot ROC curve
plt.figure()
plt.plot(FPR, TPR, marker='o', color='orange', label='ROC Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='logistic Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve-Postop day1 FDDT test')
plt.legend()
plt.show()
```

```python
# prompt: Auc

import numpy as np
from sklearn.metrics import auc

# Calculate AUC
roc_auc = auc(FPR, TPR)
print("Area under the ROC curve: {:.2f}".format(roc_auc))
```
