```python
import matplotlib.pyplot as plt

# Given values
thresholds = [0.0, 0.25, 0.5, 0.75, 1.0]
TP = [25, 22, 18, 10, 0]
FP = [25, 9, 5, 2, 0]
TN = [0, 16, 20, 23, 25]
FN = [0, 3, 7, 15, 25]

# Calculate TPR and FPR
TPR = [tp / (tp + fn) for tp, fn in zip(TP, FN)]
FPR = [fp / (fp + tn) for fp, tn in zip(FP, TN)]

# Plot ROC curve
plt.figure()
plt.plot(FPR, TPR, marker='o', color='orange', label='ROC Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='logistic Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve-Postop day1 FDDT test')
plt.legend()
plt.show()
```

```python
# prompt: Auc

import matplotlib.pyplot as plt
from sklearn.metrics import auc

# Given values
thresholds = [0.0, 0.25, 0.5, 0.75, 1.0]
TP = [25, 22, 18, 10, 0]
FP = [25, 9, 5, 2, 0]
TN = [0, 16, 20, 23, 25]
FN = [0, 3, 7, 15, 25]

# Calculate TPR and FPR
TPR = [tp / (tp + fn) if (tp+fn) else 0 for tp, fn in zip(TP, FN)]
FPR = [fp / (fp + tn) if (fp+tn) else 0 for fp, tn in zip(FP, TN)]

# Calculate AUC
roc_auc = auc(FPR, TPR)
print("AUC:", roc_auc)

# Plot ROC curve
plt.figure()
plt.plot(FPR, TPR, marker='o', color='orange', label='ROC Curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve-Postop day1 FDDT test')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import auc

# Values for Model 1 at a specific threshold (example)
TP1 = 22
FP1 = 9
TN1 = 16
FN1 = 3

# Values for Model 2 at a specific threshold (provided)
TP2 = 23
FP2 = 21
TN2 = 4
FN2 = 2

# model3
TP3 = 23
FP3 = 20
TN3 = 5
FN3 = 2

# Calculate TPR and FPR for Model 1
TPR1 = TP1 / (TP1 + FN1)  # True Positive Rate for Model 1
FPR1 = FP1 / (FP1 + TN1)  # False Positive Rate for Model 1

# Calculate TPR and FPR for Model 2
TPR2 = TP2 / (TP2 + FN2)  # True Positive Rate for Model 2
FPR2 = FP2 / (FP2 + TN2)  # False Positive Rate for Model 2

# model 3
TPR3 = TP3/(TP3+FN3)
FPR3 = FP3/(FP3+TN3)

# Plot ROC curves
plt.figure()
plt.plot([0, FPR1, 1], [0, TPR1, 1], marker='o', color='blue', label='Postop day1 FDDT test ROC Curve')
plt.plot([0, FPR2, 1], [0, TPR2, 1], marker='o', color='green', label='Postop 1month FDDT test ROC Curve')
plt.plot([0, FPR3, 1], [0, TPR3, 1], marker='o', color='red', label='Postop 3months FDDT test ROC curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Classifier (Threshold=0.5)')
plt.scatter(FPR1, TPR1, color='red')  # Highlight the point for Model 1
plt.scatter(FPR2, TPR2, color='purple')  # Highlight the point for Model 2
plt.scatter(FPR3, TPR3, color='orange')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Comparison of ROC Curves for FDDT test')
plt.legend()
plt.show()
```
