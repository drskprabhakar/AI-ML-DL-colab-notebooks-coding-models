```python
# prompt: display five images

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_dir = '/content/drive/MyDrive/macular odema 104 images/fundus images'
images = os.listdir(image_dir)[:5]  # Get the first five image filenames

fig, axes = plt.subplots(1, 5, figsize=(20, 5)) # Create a figure and subplots


for i, image_name in enumerate(images):
  image_path = os.path.join(image_dir, image_name)
  img = mpimg.imread(image_path)
  axes[i].imshow(img)
  axes[i].axis('off')  # Turn off axes for cleaner look
  axes[i].set_title(image_name)

plt.tight_layout() # Adjust spacing to prevent overlapping
plt.show()
```
